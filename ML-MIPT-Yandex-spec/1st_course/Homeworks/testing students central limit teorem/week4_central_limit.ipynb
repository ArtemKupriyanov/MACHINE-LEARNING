```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from matplotlib import rc
font = {'family': 'Verdana',
        'weight': 'normal'}
rc('font', **font)
%pylab inline
```

Выберу в качестве "любимого" степенное распределение. Оно неплохо описывает, например, зарплаты работников, частоты встречаемости слов в тексте, цены на товары и т.д. 

Плотность степенного распределения случайной величины $X$ (как реализовано в scipy.stats):
$$\Large \rho(x) = \alpha x ^ {\alpha - 1}, x \in [0,1], \alpha > 0$$

Замечу, что это отличается от другой частой формулировки: $\Large \rho(x) = С x ^ { - \alpha}$. Поэтому я вспомнил вузовский курс матана и вывел формулы для матожидания и дисперсии степеного распределения:

$$\Large E[X] = \int_{0}^{1} x \rho(x) dx =  \int_{0}^{1} x ~ \alpha ~ x ^ {\alpha - 1}dx = \alpha \int_{0}^{1} x ^ \alpha dx = \alpha \frac{x ^ {\alpha + 1}}{\alpha + 1} \bigg|_{0}^{1}  dx =  \frac{\alpha - 1}{\alpha}$$

$$\Large D[X] = \int_{0}^{1} (x - E[X])^2 \rho(x) dx = \int_{0}^{1} (x - \frac{\alpha - 1}{\alpha})^2 \rho(x) dx = \ldots =  \frac{\alpha}{(\alpha + 2)(\alpha + 1)^2}$$

Реализуем простенькие функции для подсчета матожидания и дисперсии степенного распределения.

```python
def expectation(alpha):
    return float(alpha) / (alpha + 1)
```

```python
def variance(alpha):
    return float(alpha) / (alpha + 2) / (alpha + 1) ** 2
```

Создадим случайную величину, распределенную по степенному  закону с коэффициентом $\alpha = 0.7$,  а также выборку $X$ из 1000 значений этой случайной величины. 

```python
ALPHA = 0.7
```

```python
power_rv = stats.powerlaw(a=ALPHA)
X = power_rv.rvs(size=1000)
```

Обобразим гистограмму и теоретическую плотность степенного распределения.

```python
plt.figure(figsize(10, 8))
xticks = np.linspace(0, 1, 1000)

plt.plot(xticks, power_rv.pdf(xticks))
a_hist = plt.hist(X, normed=True, bins=30)
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title(u'Гистограмма и плотность вероятности \n степенного распределения')
plt.show()
```

Порождим по 1000 подвыборок для каждого значения размера подвыборки (sample_sizes). Запишем все данные в словарь, ключами которого будут размеры подвыборок, а значениями - 1000 средних значений подвыборок соответствующего размера. 

```python
N_SAMPLES = 1000

samples_dic = {} # {sample_size: [N_SAMPLES objects]}
sample_sizes = [5, 10, 25, 50, 100, 150, 300]

for sample_size in sample_sizes:
    sample_averages = []
    for sample_id in range(N_SAMPLES):
        sample_averages.append(np.mean(power_rv.rvs(size=sample_size, random_state=sample_id)))
        
    samples_dic[sample_size] = sample_averages
```

Согласно центральной предельной теореме, чем больше число $X$ подвыборок размера $n$ мы берем, тем лучше будет выполняться:
$$\Large \bar{X}_n \sim \mathcal{N}(\mu, \frac{\sigma^2}{n}),$$
где $\bar{X}_n$ - среднее значение элементов подвыборки из $n$ элементов ($\bar{X}_n$ - тоже случайная величина), $\mu$, $\sigma$ - матожидание и дисперсия распределения случайной величины $X$ (в нашем примере - степенного распределения).

Соответственно, для степенного распределения:
    
$$\Large \bar{X}_n \sim \mathcal{N}\left(\frac{\alpha - 1}{\alpha}, \frac{\alpha}{(\alpha + 2)(\alpha + 1)^2n}\right),$$

У нас $\alpha = 0.7$, и тогда

$$\Large \bar{X}_n \sim \mathcal{N}(0.41, \frac{ 0.09}{n}).$$

Отобразим гистограммы средних значений.

```python
figsize(16,12)
f, axarr = plt.subplots(3, 2)
plt.subplots_adjust(top = 3, bottom=2)
xticks = np.linspace(0, 1, 1000)

for row in range(3):
    for col in range(2):
        n = sample_sizes[2 * row + col]
        
        
        axarr[row, col].hist(samples_dic[n], normed=True)
        axarr[row, col].set_title(u'Распределение средних для \n 1000 выборок из {0} элементов'\
                                  .format(n))
        
        norm_rv = stats.norm(loc=expectation(ALPHA), scale=sqrt(variance(ALPHA) / n))
        axarr[row, col].plot(xticks, norm_rv.pdf(xticks))
```

**Вывод**

Мы увидели, что при увеличении числа элементов в подвыборке выборки средние значения этих элементов распределяются все ближе к нормальному распределению. То есть мы на практике проверили справедливость Центральной Предельной Теоремы.
