```python
with open('wine.names.dat') as names_file:
    start = False
    features =  ['Class']
    cnt = 0
    for line in names_file:
        if '1) Alcohol' in line:
            start = True
        if start:
            features.append(line.strip().split(')')[1].strip())
            cnt += 1
        if cnt > 12:
            break
print(len(features))
```

```python
data = pd.read_csv('wine.data.dat', names=features)
data.head()
```

```python
kf_cv = KFold(data.shape[0], n_folds=5, shuffle=True, random_state=42)
```

```python
y = data['Class']
X = data.drop('Class', axis=1)
```

```python
X
```

```python
k_arr = np.linspace(1, 50, 50, dtype=int)
```

```python
scores = []
for k in k_arr:
    clf = KNeighborsClassifier(n_neighbors=k)
    scores.append(cross_val_score(clf, X, y, scoring='accuracy', cv=kf_cv).mean())
```

```python
plt.plot(k_arr, scores)
plt.xlabel('k')
plt.ylabel('score')
```

```python
max_score = -1
max_i = -1
for i in range(len(scores)):
    if scores[i] > max_score:
        max_score = scores[i]
        max_i = i
print(max_i + 1, max_score)
```
