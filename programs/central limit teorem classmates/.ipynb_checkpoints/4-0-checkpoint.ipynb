```python
import pandas as pd
import numpy as np
import scipy.stats as sts
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# select Rayleigh continuous random variable with scale = 0.5
rv_scale=0.5
rv = sts.rayleigh(scale=rv_scale)

# calculate mean/var based on wiki (https://en.wikipedia.org/wiki/Rayleigh_distribution)
rv_mean = rv_scale * np.sqrt(np.pi / 2)
rv_var = ((4 - np.pi) / 2 ) * rv_scale ** 2

print("Theory mean / var = %.4f / %.4f" % (rv_mean, rv_var))
```

```python
# Calculate hist based on random numbers
fig = plt.figure(figsize=(20,10))
ax = plt.subplot(111)

ax.hist(rv.rvs(size=1000), normed=True, bins=20)
x = np.linspace(rv.ppf(0.01),rv.ppf(0.999), 100)
ax.plot(x, rv.pdf(x), 'r-', lw=5, alpha=0.6, label='rayleigh pdf')
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.legend(loc='best')
```

```python
# calculate mean distribution based on random variable selection with different n
fig, axarr = plt.subplots(2, 2, figsize=(20, 10))

# list of selected n
nn_list = [10, 50, 100, 500]

for (i, nn) in enumerate(nn_list):
    ax = axarr[int(i/2)][int(i%2)]
    
    # generate variables and calculate mean values
    rvm = rv.rvs(size=[1000, nn]).mean(axis=1)

    # plot hist of mean
    _ = ax.hist(rvm, normed=True, bins=20, label='mean value')
    
    # generate norm distribution according Central limit theorem
    rv_norm = sts.norm(loc=rv_mean, scale=np.sqrt(rv_var/nn))
    
    # plot it!
    x = np.linspace(rv_norm.ppf(0.01),rv_norm.ppf(0.999), 100)
    ax.plot(x, rv_norm.pdf(x), 'r-', lw=5, alpha=0.6, label='theory norm pdf')
    
    # test that mean distribution is normal
    test = sts.normaltest(rvm)
    
    # print information
    ax.set_title('size: [%d], norm mean/std = %f / %f, pvalue - %.5f' % (nn, rv_mean, np.sqrt(rv_var/nn), test.pvalue))
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')
    ax.legend(loc='best')
```
